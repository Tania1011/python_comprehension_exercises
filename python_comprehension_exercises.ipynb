Using `numbers`, create a list of the **squares** of all **positive** numbers (strictly greater than 0).

### Exercise L5 — FizzBuzz with Comprehension

Using `numbers`, produce a list where:

- If divisible by **5 only**, replace it with `"Buzz"`
- Otherwise keep the **original number**

Include **all** numbers (no filtering).

**Expected output:** `['Fizz', -7, 'Fizz', 'FizzBuzz', 23, -1, 8, 'Fizz', -99, 17, 'Fizz', 56, -34, 'Buzz', 7, -2, 88, 11]`

> 💡 **Hint:** Use a **conditional expression** (ternary): `"FizzBuzz" if ... else "Fizz" if ... else "Buzz" if ... else n`. Order matters — check divisibility by 15 first.

````python
# Your answer here (List Comprehension approach)
# ── List 1: Mixed integers (some negative, some large) ──────────────────────
numbers = [4, -7, 15, 0, 23, -1, 8, 42, -99, 17, 3, 56, -34, 100, 7, -2, 88, 11]

result_L5 = ["FizzBuzz" if num % 3==0 and num %5==0 else "Fizz" if num % 3==0 else "Buzz" if num % 5==0 else num  for num in numbers]

print(result_L5)
````

### Exercise L9 — City Abbreviations

Using `cities`, create a list of **uppercase 3-letter abbreviations** for each city whose name contains **more than one word** (i.e. has a space in it). The abbreviation uses the **first letter of each of the first three tokens** (words).

Example: `"New York"` → `"NY"` (only 2 words, so just 2 letters), `"Buenos Aires"` → `"BA"`, `"São Paulo"` → `"SP"`.

Actually, make it simpler: produce a list of tuples `(city, initials)` where **initials** = the joined first-letters of each word, uppercased — for **all** cities that have a space in the name.

> 💡 **Hint:** `" ".join(...)` won't help here — try `"".join(w[0] for w in city.split())`. Add an `if " " in city` filter.

````python
# Your answer here (For-Loop approach)

result_L9 = []

for city in cities:
    # Check if the city has more than one word
    if " " in city:
        # Split the city into words
        words = city.split()

        # Create an empty string for initials
        initials = ""

        # Loop over each word
        for word in words:
            # Take the first letter of the word
            first_letter = word[0]

            # Add it to the initials string
            initials = initials + first_letter

        # Convert initials to uppercase
        initials_upper = initials.upper()

        # Append a tuple of (city, initials) to the result list
        result_L9.append( initials_upper)

# Print the final list
print(result_L9)
````

### Exercise D2 — Discounted Prices

Using `product_price`, create a new dictionary with the same keys but values **reduced by 15%**, only for products **priced above $100**. Round prices to 2 decimal places.

> 💡 **Hint:** Multiply the price by `0.85` in the value expression, and use `if price > 100` as the filter.

````python
# Your answer here (Dictionary Comprehension approach)
product_price = {
    "Laptop": 1199,
    "Mouse": 25,
    "Keyboard": 149,
    "Monitor": 301,
    "USB Cable": 10
}

discounted_prices = {
    product: round(price * 0.85, 2)              # apply 15% discount and round to 2 decimals
    for product, price in product_price.items()  # iterate over products
    if price > 100 }                             # only for items with price > 100

print(discounted_prices)
````

````python
# Your answer here (For-Loop approach)

# Initialize empty dictionary
discounted_prices = {}

# Loop over all products
for product, price in product_price.items():
    # Apply discount only if price > 100
    if price > 100:
        discounted_price = price * 0.85  # apply 15% discount
        discounted_price = round(discounted_price, 2)  # round to 2 decimals
        discounted_prices[product] = discounted_price  # add to dictionary

# Print the final dictionary
print(discounted_prices)
````

### Exercise D6 — Student Grade Categories

Using `students`, build a dictionary of `name → letter_grade` using standard grading:
- `"A"` for grade ≥ 90
- `"B"` for 80 ≤ grade < 90
- `"C"` for 70 ≤ grade < 80
- `"D"` for 60 ≤ grade < 70
- `"F"` for grade < 60

> 💡 **Hint:** Unpack tuples with `for name, grade, _ in students` (use `_` to discard the `passed` field). Chain ternaries for the grade bucket.

````python
# ── List 6: Tuples – (student_name, grade, passed) ──────────────────────────
students = [
    ("Alice",   91, True),
    ("Bob",     54, False),
    ("Carol",   78, True),
    ("Dave",    45, False),
    ("Eve",     88, True),
    ("Frank",   62, True),
    ("Grace",   35, False),
    ("Heidi",   95, True),
    ("Ivan",    71, True),
    ("Judy",    50, False),
    ("Karl",    83, True),
    ("Laura",   29, False),
]
# Your answer here
result_D6 = {}
for name, grade, _ in students:
  if grade >= 90:
    letter_grade = "A"
  elif 80 <= grade < 90:
    letter_grade = "B"
  elif 70 <= grade < 80:
    letter_grade = "C"
  elif 60 <= grade < 70:
    letter_grade = "D"
  else:      # grade < 60
    letter_grade = "F"

   # Assign the letter grade directly
    result_D6[name] = letter_grade

print(result_D6)
````

### Exercise D7 — Nested Dict Comprehension: Multiplication Table

Build a **nested dictionary** representing a 5×5 multiplication table:
`{1: {1: 1, 2: 2, ..., 5: 5}, 2: {1: 2, 2: 4, ..., 5: 10}, ...}`

> 💡 **Hint:** Use a dict comprehension whose **values** are themselves dict comprehensions:
> `{i: {j: i*j for j in range(1,6)} for i in range(1,6)}`

````python
# Your answer here
result_D7 = {}

for i in range(1, 6):              # rows
    result_D7[i] = {}              # create inner dictionary

    for j in range(1, 6):          # columns
        result_D7[i][j] = i * j    # multiply

# Step 2: Loop over the rows and print nicely
for row, cols in result_D7.items():
    print(row, "→", cols)
````

> 💡 **Hint:** Use `sentence.lower()` and filter with `if c.isalpha()`. Use `sentence.lower().count(c)` or a set of unique chars as your iterator.

````python
sentence = "The quick brown fox jumps over the lazy dog."
# Your answer here
result_D9 = {}
sentence = sentence.lower()  # all lowercase
for c in sentence:
  if c.isalpha():
        if c in result_D9:
            result_D9[c] += 1
        else:
            result_D9[c] = 1

# Print sorted by character for readability
print(dict(sorted(result_D9.items())))
````

### Exercise D10 — Department Salary Summary

Using `employees`, create a dictionary mapping each **department name** to the **average salary** of its employees (rounded to 2 decimal places).

**Expected output:** `{'Engineering': 99200.0, 'Marketing': 70000.0, 'HR': 59500.0, 'Finance': 78000.0}`

> 💡 **Hint:** First collect the unique department names. Then for each department, filter all employee records to that dept, extract salaries into a list, and compute `sum(salaries) / len(salaries)`. You can embed a list comprehension inside the dict comprehension as the value expression.

````python
# Your answer here
# Step 1: Prepare dictionaries to hold totals and counts
salary_totals = {}
employee_counts = {}

# Step 2: Loop through employees and update totals and counts
for emp_id, info in employees.items():
    dept = info["dept"]
    salary = info["salary"]

    if dept not in salary_totals:
        salary_totals[dept] = 0
        employee_counts[dept] = 0

    salary_totals[dept] += salary
    employee_counts[dept] += 1

# Step 3: Compute averages and store in result dictionary
result_D10 = {}
for dept in salary_totals:
    avg_salary = salary_totals[dept] / employee_counts[dept]
    result_D10[dept] = round(avg_salary, 2)

# Step 4: Print the result
print(result_D10)
````

````python
# ── Step 1: Get unique departments ─────────
departments = set(info["dept"] for info in employees.values())

# ── Step 2: Compute average salary for each department ─────────
result_D10 = {
    dept: round(
        sum(info["salary"] for info in employees.values() if info["dept"] == dept) /
        sum(1 for info in employees.values() if info["dept"] == dept),
        2
    )
    for dept in departments
}

# ── Step 3: Print the result ─────────
print(result_D10)
````
